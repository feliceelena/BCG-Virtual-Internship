## **Data Visualization**

### **Churn Percentage**

```python
# Data
df_churn = df_client.loc[df_client['churn'] == 1]
df_notchurn = df_client.loc[df_client['churn'] == 0]

# Create a dataframe with the data
data = {'Category': ['Churn', 'Retention'], 'Number': [len(df_churn), len(df_notchurn)]}
df = pd.DataFrame(data)

# Plot the pie chart with labels and percentages
plt.pie(df['Number'], labels=df['Category'], autopct='%1.1f%%', textprops={'fontsize': 12},
        colors = ['FireBrick', 'ForestGreen'], explode = [0, 0.1], shadow = True)
plt.title('Churn Rate')
plt.show()
```

### **Churn based on Sales Channel**

```python
# Data
channel = df_client[['id', 'channel_sales', 'churn']]
channel_2 = channel.groupby('channel_sales').count().reset_index()
channel_3 = channel.groupby('channel_sales').sum().reset_index()
channel_2 = channel_2.rename(columns={"churn": "total"})
channel_2['churn'] = channel_3['churn']
channel_2['retention'] = channel_2['total'] - channel_2['churn']

# List
churn_list = channel_2['churn'].tolist()
retention_list = channel_2['retention'].tolist()
channel_list = channel_2['channel_sales'].tolist()

# Dataframe
df = pd.DataFrame({'Churn': churn_list, 'Retention': retention_list}, index = channel_list)

# Get the totals for each row
totals = df.sum(axis=1)

# Calculate the percent for each row
percent = df.div(totals, axis=0).mul(100).round(2)

# Create the plot
ax = percent.plot(kind='barh', stacked=True, figsize=(10, 5), color=['FireBrick', 'ForestGreen'], xticks=[])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2, frameon=False)
ax.tick_params(left=False, bottom=False)
ax.spines[['top', 'bottom', 'left', 'right']].set_visible(False)
ax.set_title('Churn based on Channel Sales')

# iterate through each container
for c in ax.containers:

    # custom label calculates percent and add an empty string so 0 value bars don't have a number
    labels = [f'{w:0.2f}%' if (w := v.get_width()) > 0 else '' for v in c]

    # add annotations
    ax.bar_label(c, labels=labels, label_type='center', padding=1, color='w')
```

### **Churn based on Consumption**

```python
# Data
cons = df_client[['id', 'cons_12m','cons_gas_12m', 'cons_last_month', 'imp_cons', 'churn']]
cons_index = ['cons_12m','cons_gas_12m', 'cons_last_month', 'imp_cons']
df_churn = cons.loc[df_client['churn'] == 1]
df_notchurn = cons.loc[df_client['churn'] == 0]

churn_list = []
retention_list = []
for i in cons_index:
  total_churn = df_churn[i].sum(axis = 0)
  total_ret = df_notchurn[i].sum(axis = 0)
  churn_list.append(total_churn)
  retention_list.append(total_ret)
df = pd.DataFrame({'Churn': churn_list, 'Retention': retention_list}, index = cons_index)
df = df.div(df.sum(axis=1), axis=0)

fig, ax = plt.subplots(figsize = (12,6))
ax.bar(df.index, df["Retention"], label = "Retention", width = 0.3)
ax.bar(df.index, df["Churn"], label = "Churn", width = 0.3)

# Add labels
for c in ax.containers:
    labels = [str(round(v.get_height()*100,2)) + "%" if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c,
                 label_type='center',
                 labels = labels,
                 size = 14)
for s in ["top", "right"]:
    ax.spines[s].set_visible(False)# Add labels
ax.tick_params(labelsize = 14, labelrotation = 0)
ax.set_ylabel("Percentage", size = 14)
ax.set_xlabel("Consumption", size = 14)# Add legend
ax.legend(labels = ["Retention", "Churn"],
          fontsize = 12,
          loc = 'upper right')
ax.set_title('Churn based on Consumption')
```

```python
# Data
sns.set_theme(rc={'figure.figsize':(14,11)}) # width x height

#define plotting region (2 rows, 2 columns)
fig, axes = plt.subplots(2, 2)
plt.subplots_adjust(hspace = 0.3)

#create boxplot in each subplot
sns.boxplot(data=df_client, x='churn', y='cons_12m', ax=axes[0,0]).set(title='Cons_12m', ylabel = None)
sns.boxplot(data=df_client, x='churn', y='cons_gas_12m', ax=axes[0,1]).set(title='cons_gas_12m', ylabel = None)
sns.boxplot(data=df_client, x='churn', y='cons_last_month', ax=axes[1,0]).set(title='cons_last_month', ylabel = None)
sns.boxplot(data=df_client, x='churn', y='imp_cons', ax=axes[1,1]).set(title='imp_cons', ylabel = None)
```

### **Churn based on Category**

```python
# Data
category = df_client[['has_gas', 'churn']]
df_churn = category.loc[category['churn'] == 1]
df_notchurn = category.loc[category['churn'] == 0]

# Data Frame
churn_list = [df_churn['has_gas'].value_counts()[1], df_churn['has_gas'].value_counts()[0]]
retention_list = [df_notchurn['has_gas'].value_counts()[1], df_notchurn['has_gas'].value_counts()[0]]
cat_index = ['Electricity and Gas', 'Electricity Only']
df = pd.DataFrame({'Churn': churn_list, 'Retention': retention_list}, index = cat_index)
df = df.div(df.sum(axis=1), axis=0)

# Plot
fig, ax = plt.subplots(figsize = (7,5))
ax.bar(df.index, df["Retention"], label = "Retention")
ax.bar(df.index, df["Churn"], label = "Churn")

# Add labels
for c in ax.containers:
    labels = [str(round(v.get_height()*100,2)) + "%" if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c,
                 label_type='center',
                 labels = labels,
                 size = 14)
for s in ["top", "right"]:
    ax.spines[s].set_visible(False)# Add labels
ax.tick_params(labelsize = 14, labelrotation = 0)
ax.set_ylabel("Percentage", size = 14)
ax.set_xlabel("Category", size = 14)# Add legend
ax.legend(labels = ["Retention", "Churn"],
          fontsize = 12,
          loc = 'upper right')
ax.set_title('Churn based on Category')
```

## **Price Correlation with Churn**

```python
df_price2 = df_price_merge[['id', 'price_date','price_off_peak_var', 'price_off_peak_fix']]

# Price off_peak average for three different timespan
mean_year = df_price2.groupby('id').mean().reset_index()
mean_6m = df_price2[df_price2['price_date'] > '2015-06-01'].groupby(['id']).mean().reset_index()
mean_3m = df_price2[df_price2['price_date'] > '2015-10-01'].groupby(['id']).mean().reset_index()

# Rename columns
mean_6m = mean_6m.rename(columns={"price_off_peak_var": "price_off_peak_var_6m",
                                  "price_off_peak_fix": "price_off_peak_fix_6m"})
mean_3m = mean_3m.rename(columns={"price_off_peak_var": "price_off_peak_var_3m",
                                  "price_off_peak_fix": "price_off_peak_fix_3m"})

# Delete price_date columns on every data
mean_year = mean_year.drop(columns=['price_date'])
mean_6m = mean_6m.drop(columns=['price_date'])
mean_3m = mean_3m.drop(columns=['price_date'])

# Merge data
merge_price = pd.merge(mean_year, mean_6m, on = 'id')
merge_price = pd.merge(merge_price, mean_3m, on = 'id')
df_merge = pd.merge(merge_price, df_client[['id','churn']], on = 'id')
df_merge
```

```python
price_churn = df_merge.drop(columns=['id'])
corr = price_churn.corr()
# Plot correlation
plt.figure(figsize=(20,18))
sns.heatmap(corr, xticklabels=corr.columns.values, yticklabels=corr.columns.values, cmap='Greens', annot = True, annot_kws={'size':10})
# Axis ticks size
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

According to the correlation heatmap, it can be seen on the bottom row that price_off_peak has a very little correlation with the churn situation. Therefore, price sensitivity doesn't affect the customer churn.
